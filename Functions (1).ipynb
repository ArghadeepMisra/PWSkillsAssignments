1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list

```python
lis1 = [1,2,3,4,5]

def sum_even(numbers):
  return sum(n for n in numbers if n % 2 == 0)

print (sum_even(lis1))
```

3. 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def if_prime(n):
  if n < 2 or n > 200:
      return False

  for i in range (2, n):
      if n % i == 0:
        return False

  return True

print(if_prime(4))
print(if_prime(173))
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
num = 10

def fibo(num):
  a, b = 0, 1

  while a <= num:
    print(a)
    a, b = b, a + b

print(fibo(num))
```

6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
ex_max = 5

def pow(ex_max):
  for i in range(ex_max + 1):
    yield 2 ** i

for num in pow(ex_max):
    print(num)
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple

```python
lis4 = [(1, 4), (3, 1), (5, 8)]

sort_lis = sorted(lis4, key=lambda n: n[1])

print(sort_lis)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

```python
cel = [1, 3, 12]

fahren = list(map(lambda c: (c * 9/5) + 32, cel))

print (fahren)
```

```python
def vow_rem(str5):
  vow = "aeiouAEIOU"
  final_str = ""

  for v in str5:
    if v not in vow:
      final_str += v

  return final_str

print(vow_rem("Hello World"))
```
