# Files, exceptional handling, logging and memory management Questions

1. What is the difference between interpreted and compiled languages ?

- Compiled language like the C++ , what they do is they first converst the code that you have written to machine code and that too using the compiler. Then you have to run the executable file yourself. But interpreted Languages like Python doesn't work like that. They run step by step, preceeding by line. That's why Python or any interpreted language is slower to use but it makes it easy to code in.

2. What is exception handling in Python?

- Exception handling is when you use try except blocks to catch errors so your program doesn't crash completely. Like if user enters wrong input or file doesn't exist, instead of program stopping you can handle it nicely and show proper message.

3. What is the purpose of the finally block in exception handling?

- Finally block always runs whether error happens or not. So if you need to close a file or clean up something, you put it in finally and it will definitely execute. Even if there's return statement in try block, finally still runs.

4. What is logging in Python?

- Logging is much better than using print statements for debugging. You can save messages to files, set different levels like ERROR WARNING INFO, and control what shows in development vs production. Professional way to track what your program is doing.

5. What is the significance of the __del__ method in Python?

- __del__ is destructor method that gets called when object is about to be deleted by garbage collector. You can use it to cleanup stuff like closing files or connections. But problem is you never know exactly when it will be called.

6. What is the difference between import and from ... import in Python?

- When you do import math you have to use math.sqrt(). But if you do from math import sqrt then you can directly use sqrt(). First way keeps namespace clean, second way is more convenient but can cause name conflicts if you're not careful.

7. How can you handle multiple exceptions in Python?

- You can have multiple except blocks for different types of errors, or put multiple exception types in one except block using tuple. Also you can have general except block at end to catch any other errors you didn't think of.


8. What is the purpose of the with statement when handling files in Python?

- With statement automatically opens and closes files for you. Even if some error happens in between, file will still get closed properly. Much safer than manually opening and closing because you might forget to close file if error occurs.

9. What is the difference between multithreading and multiprocessing?

- Multithreading runs multiple threads in same process sharing memory but Python has GIL which limits it. Multiprocessing creates completely separate processes with their own memory so they can run truly parallel on multiple CPU cores. Use threading for IO stuff, multiprocessing for CPU heavy calculations.

10. What are the advantages of using logging in a program?

- Logging lets you track program behavior properly, you can save logs to files for later analysis, set different levels for different environments, and much better than scattered print statements everywhere. Also you can send logs to different places like console, files, even email for critical errors.


11. What is memory management in Python?

- Python handles memory automatically so you don't have to worry about malloc free like in C. It uses reference counting - when object has no more references it gets deleted automatically. Also has garbage collector for circular references that reference counting can't handle.

12. What are the basic steps involved in exception handling in Python?

- First you identify code that might cause problems and put it in try block. Then you add except blocks for specific errors you expect. You can also add else block for when everything goes fine, and finally block for cleanup stuff that must happen no matter what.

13. Why is memory management important in Python?

- Python does handle memory by itself but you should still know how it works so your programs don't become super slow. If you create too many big objects or don't delete stuff properly your program might run out of memory and crash.

14. What is the role of try and except in exception handling?

- Try block contains the risky code that might fail. Except blocks catch specific types of errors and decide what to do with them. So instead of your program just stopping when something goes wrong, you can handle it properly and maybe show a nice error message.

15. How does Python's garbage collection system work?

- Python keeps count of how many things are pointing to each object. When nobody is using an object anymore (count becomes zero), it gets deleted right away. But sometimes objects point to each other in circles, so there's another garbage collector that runs sometimes to clean up those circular references.

16. What is the purpose of the else block in exception handling?

- Else block only runs if no exception occurred in try block. Its useful for code that should only execute when everything went smoothly. Helps separate the normal flow from error handling logic.

17. What are the common logging levels in Python?

- There are 5 main levels: DEBUG (most detailed), INFO (general info), WARNING (something might be wrong), ERROR (definite problem), CRITICAL (very serious error). You can set minimum level so only important messages get logged in production.

18. What is the difference between os.fork() and multiprocessing in Python?

- os.fork() is Unix only system call that creates exact copy of current process. Very low level and platform specific. multiprocessing module works on all platforms (Windows, Mac, Linux) and provides easier high level interface for creating and managing processes.

19. What is the importance of closing a file in Python?

- If you don't close files they keep using system resources and you can hit "too many open files" limit. Also buffered data might not get written to disk properly. Always close files when done or better yet use with statement which handles it automatically.

20. What is the difference between file.read() and file.readline() in Python?

- So read() basically loads everything from the file into memory which is okay for small files but can be problematic if file size is too big. readline() just gets one line each time you call it so memory usage stays low. Better approach for handling large files.

21. What is the logging module in Python used for?

- Logging module gives you professional way to record what happens in your program. You can format messages nicely, send them to different destinations (console, files, network), filter by severity level, and much more organized than print debugging.

22. What is the os module in Python used for in file handling?

- Os module lets you do operating system level stuff with files like checking if file exists, creating directories, getting file permissions, listing directory contents. Basically anything file related that goes beyond just reading and writing content.

23. What are the challenges associated with memory management in Python?

- Main issues are memory leaks from circular references, keeping large objects in memory longer than needed, global variables that never get cleaned up, and using inefficient data structures. Even though Python manages memory automatically, bad programming can still cause problems.

24. How do you raise an exception manually in Python?

- You use raise keyword followed by exception type and optional message. Like raise ValueError("age cannot be negative"). Useful for validating input data or when you want to convert one type of error to more specific one that makes sense in your application.

25. Why is it important to use multithreading in certain applications?

- Multithreading prevents your application from freezing when doing slow operations like network requests or file IO. User interface stays responsive, multiple downloads can happen at same time, and overall user experience is much better compared to doing everything sequentially.

```python
file1 = open("input.txt", "r")
content = file1.read()
file1.close()

file2 = open("output.txt", "w")
file2.write(content)
file2.close()
file2 = open("output.txt", "r")

for context in file2:
  print(context)

file2.close()
```

5. How would you catch and handle division by zero error in Python?

```python
try:
    a = 10
    b = 0
    result = a/b
    print(result)
except ZeroDivisionError:
    print("cannot divide by zero")
```

6. Write a Python program that logs an error message to a log file when a division by zero exception occurs.

```python
import logging

logging.basicConfig(filename='errors.log', level=logging.ERROR)

try:
    x = 10
    y = 0
    z = x/y

except ZeroDivisionError:
    logging.error("division by zero error")
    print("error logged")
```

15. Write a Python program that prints the content of a file and handles the case when the file is empty.

```python
try:
    file1 = open("empty.txt", "r")
    content = file1.read()
    file1.close()

    if len(content) == 0:
        print("file is empty")
    else:
        print(content)
except FileNotFoundError:
    print("file not found")
```

16. Demonstrate how to use memory profiling to check the memory usage of a small program.

17. Write a Python program to create and write a list of numbers to a file, one number per line.

```python
numbers = [1, 2, 3, 4, 5]

file1 = open("numbers.txt", "w")
for num in numbers:
    file1.write(str(num) + "\n")
file1.close()

file2 = open("numbers.txt", "r")

for line in file2:
    print(line)
file2.close()
```

18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?

```python
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler('app.log', maxBytes=1024*1024, backupCount=3)
logging.basicConfig(handlers=[handler], level=logging.INFO)

for i in range(1000):
    logging.info(f"message {i}")
```

20. How would you open a file and read its contents using a context manager in Python?

```python
with open("sample.txt", "r") as file1:
    content = file1.read()
    print(content)
```

21. Write a Python program that reads a file and prints the number of occurrences of a specific word.

```python
with open("text.txt", "r") as file1:
    content = file1.read()

word = "python"
count = content.lower().count(word.lower())
print(f"'{word}' appears {count} times")
```
